```python
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By

# Initialize the WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--incognito")
driver = webdriver.Chrome(options=options)


# Facebook URL
url = "[LINK HERE]"

# Number of tabs to open in each batch
tabs_per_batch = 10

# Total iterations required
total_iterations = 2000

# Function to close login pop-up
def close_login_popup(driver):
    try:
        # Locate the close button using its position and role
        close_button = driver.find_element(By.XPATH, "//div[@aria-label='Close']")
        close_button.click()
        print("Login pop-up closed.")
    except Exception as e:
        print("No login pop-up detected or unable to close it.")

# Divide iterations into batches
batches = total_iterations // tabs_per_batch

# Start the process
for batch in range(batches):
    print(f"Starting batch {batch + 1}/{batches}")

    # Open multiple tabs in one batch
    for i in range(tabs_per_batch):
        driver.execute_script(f"window.open('{url}', '_blank');")
        # Random delay between opening tabs
        time.sleep(random.uniform(1, 3))  # Delay between 1 to 3 seconds

        # Switch to the new tab
        driver.switch_to.window(driver.window_handles[-1])
        
        # Try to close the login pop-up
        close_login_popup(driver)
        
        # Random viewing time for each tab
        watch_time = random.uniform(30, 40)  # Random time between 8 to 15 seconds
        print(f"Watching tab {i + 1}/{tabs_per_batch} for {watch_time:.2f} seconds")
        time.sleep(watch_time)

        # Close the current tab
        driver.close()
        
        # Switch back to the main tab
        driver.switch_to.window(driver.window_handles[0])
    
    # Random delay between batches
    batch_delay = random.uniform(1,5)  # Delay between 10 to 20 seconds
    print(f"Batch {batch + 1} completed. Waiting {batch_delay:.2f} seconds before next batch.")
    time.sleep(batch_delay)

# Quit the browser after all iterations are done
driver.quit()

print(driver.page_source)
print("All iterations completed!")
```
